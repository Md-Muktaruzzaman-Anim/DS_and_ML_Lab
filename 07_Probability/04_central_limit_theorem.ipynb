```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Define the population
population_size = 100000
# Let's create a skewed population (not normal)
population = np.random.exponential(scale=2, size=population_size)

# Step 2: Parameters for sampling
sample_size = 30     # n
num_samples = 1000   # How many sample means we'll take

# Step 3: Generate sample means
sample_means = []
for _ in range(num_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_means.append(np.mean(sample))

# Step 4: Plotting
plt.figure(figsize=(12,5))

# Population histogram
plt.subplot(1,2,1)
plt.hist(population, bins=50, color='skyblue', edgecolor='black')
plt.title('Population Distribution (Skewed)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Sample means histogram
plt.subplot(1,2,2)
plt.hist(sample_means, bins=30, color='salmon', edgecolor='black')
plt.title('Sampling Distribution of the Mean')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
